```python
#NominalEncoding Method
def Nominal_encoder(model):
    features=model.columns 
    for i in range(len(features)):
        model[features[i]]=model[features[i]].replace(['A+','A','A-','B','B+','B-','C','C+','C-','D','D+','F','W','WU','I'],
                                              [4,4,3.7,3,3.4,2.7,2,2.4,1.7,1,1.4,0, 0,0,0])
```

```python
X_m1 =model1_dataset.values[:, 0:11]  # INPUTS
Y_m1 =model1_dataset.values[:, 11]    #OUTPUT
```


#**Algorithm 1:** *Linear Regression*

```python
#Train Test Splitting 
x_train,x_test,y_train,y_test=train_test_split(X_m1,Y_m1,test_size=0.2,random_state=3)

def model1Algo1(mode):
    if (mode=='train'):
        LR = LinearRegression()
        LR.fit(x_train,y_train)
        y_pred=LR.predict(x_test)

        print("mean_sqrd_error is==",mean_squared_error(y_test,y_pred))
        print('Training data r-square:', LR.score(x_train,y_train))
        print('Test data r-square:', LR.score(x_test,y_test))

        print(LR.predict([[1,1,4,3.7,3.4,1,1,1,1,1,3]]))
        print(LR.predict([[4,4,4,4,4,4,4,4,4,4,4]]))
    elif (mode=='test'):
        LR=LinearRegression()
        LR.fit(X_m1,Y_m1)
        return LR
    else:
        print("Invalid args")
model1Algo1("train")
```


#**Algorithm 2:** *Polynomial Regression*

```python
#Train Test Splitting 
x_train,x_test,y_train,y_test=train_test_split(X_m1,Y_m1,test_size=0.2,random_state=3)

def model1Algo2(mode):
    if (mode=='train'):
        deg=2   #if try 3 and 4 model will over fit
        X_train_poly2=PolynomialFeatures(degree=deg).fit_transform(x_train)
        X_train_poly2=pd.DataFrame(X_train_poly2)
        X_train_poly2=X_train_poly2.drop([0],axis=1)

        PM=LinearRegression()
        PM.fit(X_train_poly2,y_train)

        print('Training data r-square:', PM.score(X_train_poly2, y_train))

        X_test_poly2=PolynomialFeatures(degree=deg).fit_transform(x_test)
        X_test_poly2=pd.DataFrame(X_test_poly2)
        X_test_poly2=X_test_poly2.drop([0],axis=1)
        y_pred=PM.predict(X_test_poly2)

        print('Test data r-square:', PM.score(X_test_poly2,y_test))
        print('Number of non-zero features:',np.sum(PM.coef_ != 0))
        print("mean_sqrd_error is:",mean_squared_error(y_test,y_pred))

        test=PolynomialFeatures(degree=2).fit_transform([[4,2.7,1,3,1,2,1,3.7,4,4,1]])
        test=pd.DataFrame(test)
        test=test.drop([0],axis=1)
        print(PM.predict(test))
    elif (mode=='test'):
        deg=2   #if try 3 and 4 model will over fit
        modified_x=PolynomialFeatures(degree=deg).fit_transform(X_m1)
        modified_x=pd.DataFrame(modified_x)
        modified_x=modified_x.drop([0],axis=1)
        PM=LinearRegression()
        PM.fit(modified_x,Y_m1)
        return PM
    else:
        print("Invalid args")
model1Algo2("train")
```

```python
X_m2 =model2_dataset.values[:, 0:22]  # INPUTS
Y_m2 =model2_dataset.values[:, 22]    #OUTPUT
```


#**Algorithm 1:** *Linear Regression*

```python
#Train Test Splitting 
x_train,x_test,y_train,y_test=train_test_split(X_m2,Y_m2,test_size=0.2,random_state=2)
def model2Algo1(mode):
    if (mode=='train'):
        LReg = LinearRegression()
        LReg.fit(x_train,y_train)
        y_pred=LReg.predict(x_test)

        print("mean_sqrd_error is==",mean_squared_error(y_test,y_pred))
        print('Training data r-square:', LReg.score(x_train,y_train))
        print('Test data r-square:', LReg.score(x_test,y_test))

        print(LReg.predict([[4,3.6,4,2.6,2.3,4,2.6,4,1.9,4,4,2.2,4,3.4,4,2.4,3.0,4,3.7,4,3.7,1.9]]))
    elif (mode=='test'): # testing our implemented model on run time entry by user 
        LReg = LinearRegression()
        LReg.fit(X_m2,Y_m2)
        return LReg
    else:
        print('invalid args')
model2Algo1('train')
```


#**Algorithm 2:** *KNN Regressor*

```python
#Train Test Splitting 
x_train,x_test,y_train,y_test=train_test_split(X_m2,Y_m2,test_size=0.2,random_state=2)
def model2Algo2(mode):
    if mode=='train':
        params = {'n_neighbors':[2,3,9,8,7]}
        knn = neighbors.KNeighborsRegressor()
        model_knn = GridSearchCV(knn, params, cv=5)
        model_knn.fit(x_train,y_train)
        y_pred=model_knn.predict(x_test)
        print("Training Accuracy:",model_knn.score(x_train,y_train))
        print("Testing Accuracy:",model_knn.score(x_test,y_test))
        print('Mean Squared error:',mean_squared_error(y_test,y_pred))
        print(model_knn.best_params_) # this k value will be used for run time prediction prediction
        print(model_knn.predict([[4,3.6,4,2.6,2.3,4,2.6,4,1.9,4,4,2.2,4,3.4,4,2.4,3.0,4,3.7,4,3.7,1.9]]))
    elif mode=='test':
        knn=KNeighborsRegressor(n_neighbors=9) # As found during training
        model_knn=knn.fit(X_m2,Y_m2)
        return model_knn
    else:
        print("Invalid args")
model2Algo2('train')
```


#**Algorithm 1:** *Linear Regression*

```python
#Train test splitting
x_train,x_test,y_train,y_test=train_test_split(X_m3,Y_m3,test_size=0.2,random_state=3)
def model3Algo1(mode):
    if mode==('train'):
        LReg3 = LinearRegression()
        LReg3.fit(x_train,y_train)
        y_pred=LReg3.predict(x_test)

        print("mean_sqrd_error is==",mean_squared_error(y_test,y_pred))
        print('Training data r-square:', LReg3.score(x_train,y_train))
        print('Test data r-square:', LReg3.score(x_test,y_test))
        print(LReg3.predict([[4,2.7,1,3,1,2,1,3.7,4,4,1,1,1,1,4,3,2,2,2.3,2.6,1,1,1,4,4,4,4,0,0,0,2.5,3.5,4]]))
    elif mode==('test'):
        LReg3 = LinearRegression()
        LReg3.fit(X_m3,Y_m3)
        return LReg3
    else:
        print('Invalid args')
model3Algo1('train')
```


#**Algorithm 2:** *Gradient Boosting Regressor*

```python
#train test splitting
x_train,x_test,y_train,y_test=train_test_split(X_m3,Y_m3,test_size=0.2,random_state=3)

# these parameters are best for my model, have used hit and trial to obtain these.
params = {
    "n_estimators": 500,
    "max_depth": 4,
    "min_samples_split":4,
    "learning_rate": 0.1,   # tried different learning rates found it best
    "loss": "squared_error",
}
def model3Algo2(mode):# Here we have changed each our model training in function so that when user calls our algo we test it on entire sample 
    if mode=='train':
        reg_model = ensemble.GradientBoostingRegressor(**params)
        reg_model.fit(x_train, y_train)
        mse = mean_squared_error(y_test, reg_model.predict(x_test))
        print("The mean squared error on test set is:", mse)
        print("Training Accuracy:",reg_model.score(x_train,y_train))
        print("Testing Accuracy:",reg_model.score(x_test,y_test))  

        print(reg_model.predict([[4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4]]))
        print(reg_model.predict([[4,2.7,1,3,1,2,1,3.7,4,4,1,1,1,1,4,3,2,2,2.3,2.6,1,1,1,4,4,4,4,0,0,0,2.5,3.5,4]]))
    elif mode=='test':
        reg_model = ensemble.GradientBoostingRegressor(**params)
        reg_model.fit(X_m3,Y_m3)
        return reg_model
    else:
        print('wrong args')
model3Algo2('train')
```
